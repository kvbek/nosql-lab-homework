# Zadanie 4 — Redis: key-value i TTL

```python
import redis

r = redis.Redis(host='redis_nosql_lab', port=6379, decode_responses=True)
r.ping()
print('Polaczono z Redisem')
```

## Polecenie 1 — Zapis danych

```python
r.set('user:1', 'Ala')
r.set('user:2', 'Bartek')
r.set('user:3', 'Cezary')
print('Zapisano: user:1 -> Ala, user:2 -> Bartek, user:3 -> Cezary')
```

## Polecenie 2 — Odczyt klucza user:1

```python
value = r.get('user:1')
print(f'user:1 -> {value}')
```

## Polecenie 3 — Ustawienie TTL dla user:1 (60 sekund)

```python
r.expire('user:1', 60)
print('Ustawiono TTL = 60s dla user:1')
```

## Polecenie 4 — Sprawdzenie pozostałego czasu do wygasniecia

```python
ttl = r.ttl('user:1')
print(f'Pozostaly czas TTL dla user:1: {ttl} sekund')
```

---
## Co stanie się z kluczem po upływie TTL?

Po upływie 60 sekund Redis automatycznie usunie klucz `user:1` — próba odczytu przez `GET user:1` zwróci `None` (klucz nie istnieje). Mechanizm TTL służy do automatycznego czyszczenia danych tymczasowych bez konieczności ręcznego usuwania kluczy.
